# 2 Explain the difference between mutable and immutable data types with examples
- Mutable Data Types:
Mutable data types are those whose internal state or content can be modified after they are created. Operations performed on mutable objects can change the object's value in memory without creating a new object.
Examples: Lists, Dictionaries, Sets (in Python).
Characteristic: Allows in-place modification.
Use Case: Suitable for data that needs frequent updates or dynamic changes.

Immutable Data Types:
Immutable data types are those whose internal state or content cannot be modified after they are created. Any operation that appears to change an immutable object actually results in the creation of a new object with the modified value, leaving the original object unchanged.
Examples: Strings, Tuples, Integers, Floats (in Python).
Characteristic: Cannot be changed after creation; any "modification" creates a new object.
Use Case: Beneficial for data integrity, thread safety (in multi-threaded environments), and as keys in hash-based data structures (like dictionary keys or set elements).

#4 Describe how dictionaries store data
-  Dictionaries store data as key-value pairs.
Each key must be unique and immutable, and it maps to a specific value. This allows for efficient data retrieval using the key.

#5 Why might you use a set instead of a list in Python?
- Sets are used when the presence of a value needs to be checked quickly, or when duplicate values are not allowed.
Sets have faster membership testing (using the "in" keyword) and automatically handle duplicate removal, which lists do not.

#6 What is a string in Python, and how is it different from a list?
- A string is an immutable sequence of characters, while a list is a mutable sequence of objects.
Strings are specifically for text data and are immutable, meaning individual characters cannot be changed in place. Lists can hold various data types and can be modified after creation.

#7 How do tuples ensure data integrity in Python?
- Tuples ensure data integrity because they are immutable.
Once a tuple is created, its elements cannot be changed, added, or removed, which prevents accidental modification of the data they hold.

#8 . What is a hash table, and how does it relate to dictionaries in Python?
- A hash table is a data structure that implements an associative array abstract data type, and Python dictionaries are implemented using hash tables.
Hash tables use a hash function to compute an index into an array of buckets or slots, from which the desired value can be found. This provides very fast average-case time complexity for insertion, deletion, and access operations.

#9 Can lists contain different data types in Python?
- Yes, lists in Python can contain different data types.
Python lists are designed to be flexible and can store integers, strings, floats, other lists, or any combination of these.

#10 . Explain why strings are immutable in Python
- Strings are immutable in Python primarily for performance optimization and consistency.
Immutability allows strings to be used as keys in dictionaries (which require immutable keys) and makes string operations more predictable. When a string is "modified", a new string object is actually created.

#11 What advantages do dictionaries offer over lists for certain tasks?
- Dictionaries offer faster data retrieval based on a specific key.
For tasks involving mapping data or looking up values by a unique identifier (like a name or ID), dictionaries provide average constant time complexity, which is much faster than searching through a list.

#12 Describe a scenario where using a tuple would be preferable over a list.

#13 How do sets handle duplicate values in Python?
- Sets automatically discard duplicate values.
When you add a duplicate element to a set or convert a list with duplicates to a set, the set will only store one instance of that value.

#14 How does the "in" keyword work differently for lists and dictionaries?
- For lists, the "in" keyword checks for the presence of a value among the elements, while for dictionaries, it checks for the presence of a key.
Checking membership in a dictionary is generally much faster than in a list.

#18 In what situations are lists preferred over dictionaries?

#19. Why are dictionaries considered unordered, and how does that affect data retrieval?

In modern Python, dictionaries maintain insertion order, but they are still fundamentally designed for key-based access, not positional access like lists.
